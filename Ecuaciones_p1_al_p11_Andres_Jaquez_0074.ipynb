### Ecuaciones de la tabla  1
 $x + 8 = 15$
 #### Restamos 8 en ambos lados para despejar
 $x$:$x = 15 - 8$
 #### Resultado: $x = 7$



```python
from sympy import symbols, Eq, solve

x = symbols('x')
ecuacion = Eq(x + 8, 15)
solucion = solve(ecuacion, x)
print(f"Resultado de la ecuacion 1 es igual a: {solucion[0]}")
```

### Ecuacion 3

$3x + 4x = 21$
#### Agrupamos términos semejantes
($3x$ y $4x$):$7x = 21$
#### Dividimos entre 7:
$x = 21 / 7$
#### Resultado:
$x = 3$

```python
from sympy import symbols, Eq, solve

x = symbols('x')
# Agrupamos 3x + 4x internamente
ecuacion = Eq(3*x + 4*x, 21)
solucion = solve(ecuacion, x)
print(f"Resultado de la ecuacion 3 es igual a : {solucion[0]}")
```

### Ecuacion 5

$6a - 2a + a = 20$
#### Agrupamos términos semejantes
($6 - 2 + 1$):$5a = 20$
#### Dividimos entre 5:
$a = 20 / 5$
#### Resultado:
$a = 4$

```python
from sympy import symbols, Eq, solve

a = symbols('a')
# 6a - 2a + a es interpretado automáticamente
ecuacion = Eq(6*a - 2*a + a, 20)
solucion = solve(ecuacion, a)
print(f"Resultado de la ecuacion 5 es igual a : {solucion[0]}")
```

### Ecuacion 7

$4x + 3 = 2x + 11$
#### Restamos $2x$ en ambos lados:
$2x + 3 = 11$
#### Restamos 3 en ambos lados:
$2x = 8$
####Dividimos entre 2:
$x = 8/2$
#### Resultado:
$x = 4$

```python
from sympy import symbols, Eq, solve

x = symbols('x')
ecuacion = Eq(4*x + 3, 2*x + 11)
solucion = solve(ecuacion, x)
print(f"Resultado de la ecuacion 7 es igual a : {solucion[0]}")
```

### Ecuacion 9

$9m + 4 = 5m + 20$
#### Restamos $5m$ en ambos lados:
$4m + 4 = 20$
### Restamos 4 en ambos lados:
$4m = 16$
#### Dividimos entre 4:
$m = 16/4$
#### Resultado
$m = 4$

```python
from sympy import symbols, Eq, solve

m = symbols('m')
ecuacion = Eq(9*m + 4, 5*m + 20)
solucion = solve(ecuacion, m)
print(f"Resultado de la ecuacion 9 es igual a : {solucion[0]}")
```
