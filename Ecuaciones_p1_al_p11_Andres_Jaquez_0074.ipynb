##Problema 1: $-8 + 3 = -5

Análisis: Imagina que debes 8 dólares (-8) y pagas 3 (+3).

Proceso: Como los signos son diferentes, restas los valores ($8 - 3 = 5$).   

Resultado: El signo del número mayor (el 8) es negativo, por lo que el

resultado es $-5$.

```python
# Definimos los valores con sus respectivos signos
#valor_a = 3
#valor_b = -8
valor_a = int(input("Ingrese el primer valor: "))
valor_b = int(input("Ingrese el segundo valor: "))

# Realizamos la suma directa
resultado = valor_a - valor_b

# Mostramos el resultado en pantalla
print("El resultado de -8 + 3 es: ", resultado)
```

##Problema 3:
 -5 + (-7) = -12

Análisis: Tienes una deuda de 5 y le sumas otra deuda de 7.

Proceso: Al ser signos iguales, los valores se suman (5 + 7 = 12).

Resultado: Se mantiene el signo de la operación original,

resultando en $-12$.

```python
# Sumar un número negativo se escribe con paréntesis para mayor claridad
#resultado = -5 + (-7)
resultado = int(input("Ingrese el primer valor: ")) + int(input("Ingrese el segundo valor: "))

# Python interpreta +(-) como una resta
print("El resultado de -5 + (-7) es: -", resultado)
```

##Problema 5:
 $-3 - 4 = -7

 Análisis: Estás en la posición $-3$ de la recta numérica y te mueves 4 unidades más a la izquierda.

 Proceso: Es una acumulación de negativos (3 + 4 = 7).

 Resultado: El resultado final es $-7$.

```python
# Resta de un número positivo a un número negativo
# resultado = -3 - 4
resultado = int(input("Ingrese el primer valor: ")) + int(input("Ingrese el segundo valor: "))

print("El resultado de -3 - 4 es: -", resultado)
```

##Problema 7:
 -7 + 12 = 5

 * Explicación: Estás en el punto $-7$ y avanzas $12$ unidades hacia la derecha. Como el avance (12) es mayor que la posición negativa inicial (7),

 * logras cruzar el cero y llegar al lado positivo. La diferencia entre $12$ y $7$ es $5$, y es positivo porque el $12$ tiene mayor peso.

```python
# Suma de un positivo mayor a un negativo
#resultado = -7 + 12
resultado = int(input("Ingrese el primer valor: ")) - int(input("Ingrese el segundo valor: "))

print("El resultado de -7 + 12 es:", resultado)
```

##Problema 9:
$9 - (-3) = 12

Explicación: Este caso es especial por la "ley de los signos". Cuando restas un número negativo, los dos signos de menos se cancelan entre sí y se convierten en una suma (- \times - = +).

Por lo tanto, la operación se transforma mentalmente en $9 + 3$,

lo que nos da $12$.

```python
# Restar un negativo en Python genera una suma automática
#resultado = 9 - (-3)
resultado = int(input("Ingrese el primer valor: ")) + int(input("Ingrese el segundo valor: "))

print("El resultado de 9 - (-3) es:", resultado)
```
